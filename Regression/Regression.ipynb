### 1. Create a regression model between TotalWorkingYears and MonthlyIncome (response variable). Show the scatter plot with regression line between them

```python
model1 = ols("MonthlyIncome ~ TotalWorkingYears", data = df).fit()
model1.summary()
```

Since the p-value is less than 0.05, it shows that there exists a positivr linear relationship between the two variables, TotalWorkingYears and MonthlyIncome. They are directly proportional to each other as shown in the graph below. Moreover, as we know as lower the p-value, the stronger there is a relationship thus, there is a strong relationship between them which is also shown in the graph below as the points are closer to the regression line. As the TotalWorkingYears increases, the MonthlyIncome of the employees also increase. The linear equation for this model is as below:

x = TotalWorkingYears

y = MonthlyIncome

y = 1228 + 468x

468 MonthlyIncome increases for every year you work in firm. (Assuming the MonthlyIncome is in dollars($))

In addition to this, we can see that R-squared is 0.5 which is in between 0 and 1 which means that the error calculated is not so great so we can relate that the two variables are related to each other with less error.

```python
#Seaborn
sns.lmplot(x='TotalWorkingYears', y='MonthlyIncome', data=df)
```

### 4. Create a regression model to predict MonthlyIncome using all other inputs. Discuss the effectiveness of the model. Report 3 most significant inputs and 3 least significant inputs and interpret the results. Create a new input and show the prediction of MonthlyIncome using the same model.

```python
df = df.drop(['MaritalStatus'],axis=1)
```

One-Hot Encoding

```python
one_hot = pd.get_dummies(df, columns = ['Department', 'EducationField', 'Gender', 'JobRole', 'OverTime'])
```

```python
one_hot
```

```python
X = one_hot.drop(['MonthlyIncome'],axis=1)
Y = one_hot['MonthlyIncome']
```

```python
modelcat = sm.OLS(Y,sm.add_constant(X)).fit()
```

```python
modelcat.summary()
```

The most significant and insignificant inputs are decided by looking at the p-values of MonthlyIncome against all the other input variables. All the p-values below 0.05 show that there exists a strong relationship between the two variables compared and vice versa.

Three most significant inputs:


We had to look for the three least p-values to show the most significant inputs in the regression model created for MonthlyIncome. Therefore, JobLevel, TotalWorkingYears and JobRole(it includes its other subfields as shown in the table above after one-hot encoding) are the most significant inputs in linear regression model as these have the lowest p-values (0.0) against MonthlyIncome which establishes that they have the most influence over MonthlyIncome calculations for the employees.

Three least significant inputs:

We had to look for the three greatest p-values to show the most significant inputs in the regression model created for MonthlyIncome. Therefore, PerformanceRating, YearsInCurrentRole, WorkLifeBalance are the least siginificant inputs in linear regression model as these have the greatest p-values against MonthlyIncome which establishes that these variables have the least influence over MonthlyIncome calculations for the employees.

One-Hot Encoding

```python
one_hot6 = pd.get_dummies(df_6, columns = ['JobRole'])
```

```python
X6 = one_hot6.drop(['MonthlyIncome'],axis=1)
Y6 = one_hot6['MonthlyIncome']
```

```python
modelcat6 = sm.OLS(Y6,sm.add_constant(X6)).fit()
```

```python
modelcat6.summary()
```
